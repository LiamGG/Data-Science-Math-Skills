# Exponents
$3^1 = 3$

$3^2 = 3 \cdot 3$

$3^3 = 3\cdot 3\cdot 3$

$x^0 = 1$

$(x^a)^b = x^{a \cdot b}$

$x^a \cdot x^b = x^{a + b}$

$x^{-a} = \frac{1}{x^a}$

$\frac{x^a}{x^b} = x^{a-b}$

$(xy)^a = x^a\cdot y^a$

$(\frac{x}{y})^a = \frac{x^a}{y^a}$

$x^{\frac{a}{b}} = \sqrt[b]{x^a}$

# Logarithms
$b^y = x$ -> $log_{b}(x) = y$

$2^4 = 16$ -> $log_{2}(16) = 4$

$b^{log_{b}(a)} = a$

### Base change rule
$log_{a}(b) = \frac{log_{x}(b)}{log_{x}(a)}$

### Base switch rule
$log_{b}(c) = \frac{1}{log_{c}(b)}$

### Product Rule
$log_{b}(x \cdot y) = log_{b}(x) + log_{b}(y)$

### Quotient Rule
$log_{b}(\frac{x}{y}) = log_{b}(x) - log_{b}(y)$

### Power Rule
$log_{b}(x^y) = y \cdot log_{b}(x)$

$log_{b}{\sqrt{x}} = log_{b}({x^{\frac{1}{2}}}) = \frac{1}{2}log_{b}(x)$

### Derivate
$f(x) = log_{b}(c)$ -> $f'(x) = \frac{1}{x\cdot ln(b)}$

### Integral
$\int log_{b}(x) dx = x\cdot (log_{b}(x) - \frac{1}{ln(b)}) + C$

### Negative Numbers
log_{b}(x) is undefined when $X <= 0$

### Logarithm of 0
$log_{b}(0)$ is *undefined*

$lim_{x -> 0^{+}}log_{b}(x) = -\infty$

### Logarithm of 1
$log_{b}(1) = 0$


### Logarithm of the base
$log_{b}(b) = 1$


### Logarithm of infinity
$log_{b}(x) = \infty$ as $x -> \infty$

# Examples

$log_{2}(\frac{39x}{(x-5)} = 4$

$2^{log_{2}(\frac{39x}{(x-5)})} = 2^4$

$\frac{39x}{(x-5)} = 16$

$39x = 16x - 80$

$23x = -80$

$x = \frac{-80}{23}$

# Rate of Growth of Processes

### Discrete rate of growth:
$\$1.00(1 + r)^t$, $r = 100\%\; /\; year$,  $t = 1  year$

After 1 year ->$2.00. That's a 100\% discrete exponential rate of growth per year

After 2 years ->$4.00. That's a 100\% discrete exponential rate of growth per year

### Continuous rate of growth:
Elephant of 200kg, grows continuously as 5% / year. 

After x years,it will weigh $(200kg)e^{(0.05) * x}$



### Rate of return

Start with 10kg increasing as 200% / year.

How long until they weight $5.972 \times 10^24 kg$?

$5.972 \times 10^{24} kg = 10kg \cdot e^{2t}$

$5.972 \times 10^{23} kg = e^{2t}$


$ln(5.972 \times 10^{23}) = 2t$

$t = 27.37 years$
